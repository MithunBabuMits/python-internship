```python
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset('iris')

```

```python
print(iris.head())
```

**Create Bar Chart**

```python
avg_petal_width = iris.groupby('species')['petal_width'].mean()

plt.figure(figsize=(8, 6))
avg_petal_width.plot(kind='bar', color='skyblue')
plt.title('Average Petal Width of Iris Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Width')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Create Line Chart**

```python
plt.figure(figsize=(8, 6))
plt.plot(iris.index, iris['sepal_length'], marker='o', linestyle='-', color='green')
plt.title('Sepal Length in Iris Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

** Create Pie Chart**

```python
species_counts = iris['species'].value_counts()


plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightskyblue', 'lightcoral'])
plt.title('Distribution of Flower Species in Iris Dataset')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
